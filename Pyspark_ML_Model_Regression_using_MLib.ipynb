```python
house_df = sqlContext.read.format('com.databricks.spark.csv').options(header='true',inferschema='true').load('03_Boston.csv')
```

```python
#descriptive analysis
house_df.toPandas()
```

```python
#transformation (ETL)


from pyspark.ml.feature import VectorAssembler
vectorAssembler = VectorAssembler(inputCols=['crim','zn','indus','chas','nox','rm','age','dis','rad','tax','ptratio','black','lstat'],outputCol = 'features')
vhouse_df = vectorAssembler.transform(house_df)
vhouse_df.show(3)
```

```python
vhouse_df.select(['features','medv']).show(3)
vhouse_df.show()
```

```python
splits = vhouse_df.randomSplit([0.7,0.3])
type(splits)
splits[0].show
train_df = splits[0]
test_df =splits[1]

```

```python
from pyspark.ml.regression import LinearRegression
lr = LinearRegression(featuresCol='features',labelCol='medv',maxIter=10, regParam=0.3,elasticNetParam=0.8)
lr_model = lr.fit(train_df)
print("coefficients: "+str(lr_model.coefficients))
print("Intercept: " +str(lr_model.intercept))

```

```python
trainingSummary = lr_model.summary
print("RMSE: %f" % trainingSummary.rootMeanSquaredError)
print("r2: %f" % trainingSummary.r2)
```
